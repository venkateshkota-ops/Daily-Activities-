# Binning

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    "age": [18, 22, 25, 30, 35, 40, 50]
})

print(data)
```

# Equal-Width

```python
data["age_bin"] = pd.cut(data["age"], bins=4)

data
```

```python
plt.hist(data["age"], bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Width Binning using pd.cut()")
plt.show()
```

# Equal-Frequency

```python
data2 = pd.DataFrame({
    "age": [18, 22, 25, 30, 35, 40, 50]
})

print(data2)
```

```python
data2["age_frequency_bin"] = pd.qcut(data2["age"], q=4)

data2
```

# Custom/domain-based binning

```python
custom_bins = [17, 25, 35, 50]
labels = ["Young", "Adult", "Senior"]

bins = pd.cut(data["age"], bins=custom_bins, labels=labels)
bins
```

```python
plt.hist(bins, bins=4)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Equal-Frequency Binning using pd.qcut()")
plt.show()
```
